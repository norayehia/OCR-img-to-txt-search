```python
import os
from PIL import Image
import pytesseract
import shutil

# مسار مجلد الصور
source_folder = "/content/drive/MyDrive/NMU/analyses/imgsearch/Done_aie111-11212024093918"  # مجلد الصور
target_folder = "/content/drive/MyDrive/NMU/analyses/imgsearch/ms12_folder2/CSE261"  # المجلد العام لنقل الصور

# قائمة بأرقام الطلاب التي تبحث عنها
target_student_ids = ["222101512"]

# مسار Tesseract لـ Linux
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

def search_and_move_images(source, target, student_ids):
    """
    Searches for images containing specific student IDs and moves them to a target folder.
    Stops searching for an ID after finding one matching image.

    Args:
        source (str): The path to the source folder containing the images.
        target (str): The path to the target folder where the images will be moved.
        student_ids (list): List of student IDs to search for in the images.
    """
    # التأكد من وجود مجلد الوجهة
    if not os.path.exists(target):
        os.makedirs(target)
        print(f"Created target directory: {target}")

    # التأكد من وجود مجلد فرعي لكل طالب
    for student_id in student_ids:
        student_folder = os.path.join(target, f"student_{student_id}")
        if not os.path.exists(student_folder):
            os.makedirs(student_folder)

    # إعداد قائمة لتتبع الطلاب الذين تم العثور على صورهم
    found_ids = set()

    # البحث في كل الصور
    for filename in os.listdir(source):
        if filename.endswith(".tif") or filename.endswith(".png"):
            img_path = os.path.join(source, filename)

            try:
                # استخراج النص من الصورة
                text = pytesseract.image_to_string(Image.open(img_path), lang="ara+eng")

                # البحث عن كل رقم طالب في النص
                for student_id in student_ids:
                    if student_id in text and student_id not in found_ids:
                        student_folder = os.path.join(target, f"student_{student_id}")
                        print(f"تم العثور على الطالب {student_id} في الصورة: {filename}")
                        shutil.move(img_path, os.path.join(student_folder, filename))
                        print(f"تم نقل الصورة إلى {student_folder}")
                        found_ids.add(student_id)  # إضافة ID للطلاب الذين تم العثور عليهم

                        # التحقق مما إذا تم العثور على جميع IDs
                        if len(found_ids) == len(student_ids):
                            print("تم العثور على صورة لكل الطلاب. إيقاف البحث.")
                            return  # الخروج من الدالة إذا تم العثور على جميع الصور

            except Exception as e:
                print(f"Error processing {filename}: {e}")

# تشغيل الوظيفة
search_and_move_images(source_folder, target_folder, target_student_ids)
```
